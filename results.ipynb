```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import sin, cos, pi

def diffract(theta, w, s, N, lmbda):
    k = (2 * pi) / lmbda
    v = 1/2 * k * s * sin(theta)
    u = 1/2 * k * w * sin(theta)
    y = (w ** 2) * (((sin(u) / u) * (sin(N * v) / sin(v))) ** 2)
    return y
```

```python
theta  = np.linspace(-0.005, 0.005, 50)
theory = diffract(theta, 5, 0.001, 1, 650e-9)
plt.plot(sin(theta), theory)
```

```python
# code for red 0.08mm single
x,y = np.loadtxt("red1single.txt", dtype = 'float', delimiter = ",", skiprows = 10, unpack = True)
x_shift = x * (54.74e-6)
L = 1 #length of optical path
theta = np.arctan(x_shift/L)
sinTheta = sin(theta)


sinTheta_min = []
array_min = []

i = 0
while i < len(y) - 2:
    if (y[i+1] > y[i] and i != 0 and y[i - 1] > y[i]) or y[i] == 0:
        array_min.append(y[i])
        sinTheta_min.append(sinTheta[i])
    i += 1

print("sin theta: ", sinTheta_min)
print("log intensity", array_min)    

#log intensity graph
plt.figure(figsize = (10, 3))
plt.semilogy(sinTheta, y)
plt.vlines(0, 750, 0, colors = 'r', linestyle = 'dashed')
plt.xlim([-0.07, 0.07])

plt.show()


#normal graph
plt.figure(figsize = (10,10))
plt.plot(sinTheta, y, lw = 0.5)
plt.show()
```
