```python
import numpy as np
import matplotlib.pyplot as plt
a = np.genfromtxt("green1single.txt",skip_header=10,delimiter =",")

from numpy import sin, cos, pi
```

```python
x = a[:,0] - a[813,0]
Intensity = a[:,1]

```

```python
print(np.where(Intensity == np.max(Intensity)), len(x))
```

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.plot(x,Intensity)

ax.plot()
```

```python
def diffract(theta, w, s, N, lmbda):
    k = (2 * pi) / lmbda
    v = 1/2 * k * s * sin(theta)
    u = 1/2 * k * w * cos(theta)
    y = (w ** 2) * (((sin(u) / u) * (sin(N * v) / sin(v))) ** 2)
    return y
```

```python
Theory = diffract(theta,5,0.001,1,650e-9)
```
